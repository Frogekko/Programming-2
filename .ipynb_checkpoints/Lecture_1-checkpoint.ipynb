# Load Iris

```python
from sklearn import datasets

iris = datasets.load_iris()

#print(iris)
```

# Convert the data to pandas dataframe

```python
panda_data = pd.DataFrame(data = np.c_[iris['data'], iris['target']], columns = iris['feature_names'] + ['target'])
```

```python
panda_data
```

```python
panda_data.head()
```

```python
panda_data.tail()
```

```python
panda_data.describe()
```

```python
panda_data["petal length (cm)"]
```

```python
panda_data[50:61]
```

```python
panda_data.iloc[50:61, 0:2]
```

# Oppgave 3

```python
panda_data.iloc[5:16, 0:4:3]
```

```python
panda_data.at[25, 'sepal width (cm)']
```

```python
panda_data.loc[panda_data['target'] == 0].describe(include = 'all')
```

```python
import matplotlib.pyplot as plt

panda_data.hist()
plt.show()
```

```python
panda_data['petal length (cm)'].hist()
plt.show()
```

```python
# Select rows of a certain class
x1 = panda_data.loc[panda_data['target'] == 0]
x2 = panda_data.loc[panda_data['target'] == 1]
x3 = panda_data.loc[panda_data['target'] == 2]

# Possible to modify bin size as well.
plt.hist(x1['petal width (cm)'], color = 'b', label = '0')
plt.hist(x2['petal width (cm)'], color = 'r', label = '1')
plt.hist(x3['petal width (cm)'], color = 'g', label = '2')
plt.legend();
```

```python
import seaborn as sns
# Seaborn has better visualizations
sns.pairplot(panda_data)
```
